```python
import requests
import bs4
from bs4 import BeautifulSoup
import re
```

```python
from ipynb.fs.full.symptoms_scraper import get_symptoms
```

```python
def get_links(URL):
    
    mayoclinic = 'https://www.mayoclinic.org'

    r = requests.get(URL)
    soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib

    content = soup.find("div", {"class": "index content-within"})
    li = content.find_all("li")

    links = []
    for link in li:
        disease = mayoclinic + link.find('a')['href']
        links.append(disease) if disease not in links else links
        
    return links
```

```python
diseases_link = 'https://www.mayoclinic.org/diseases-conditions/index?letter='

all_links = []

from string import ascii_uppercase
for c in ascii_uppercase:
    c_links = get_links(diseases_link + c)
    all_links.extend(c_links)
    
    print(c + ": " + str(len(c_links)))
    
```

```python
all_links.extend(get_links(diseases_link + '0'))
```
